## Question 3
Create a new column 'Has_Cabin' that tells you whether passengers had a cabin or not. 

```python
df1['Has_Cabin'] = np.where(df1['Cabin'].isnull(), 'no', 'yes')
df1
```

## Question 4
Creating a new column â€œTitleâ€, which will contain the titles of the passengers such as Mr., Mrs., etc. There should be five titles: Mr, Mrs, Master, Miss, and Special. 

```python
import re
df1["Title"]=df1.Name.apply(lambda x : re.search('([A-Z][a-z]+)\.',x).group(1))
df1["Title"].unique()
```

```python
list = ["Don","Rev","Dr","Mme","Ms","Major","Lady","Sir","Mlle","Col","Capt","Countess","Jonkheer"]
df1["Title"] = df1["Title"].replace(list,"Special")
```

```python
df1["Title"].unique()
```

```python
df1
```

## Question 5
Imputing Missing Values

```python
df1=df1.fillna(df1.mean())
df1.mode()
```

```python
df2 = df1["Cabin"].fillna("B96")
df3 = df1["Embarked"].fillna("S")
df1 = df1.drop(["Cabin","Embarked"],axis =1)
```

```python
df1.isnull().sum()
```

```python
df1.insert(loc =1 ,column = "Cabin",value = df2)
df1.insert(loc =1 ,column = "Embarked",value = df3)
df1
```

```python
df1.isnull().sum()
```

## Question 6

```python
df1["Categorical_age"] = pd.qcut(df1.Age,q= 4,labels = False)
df1["Categorical_fare"] = pd.qcut(df1.Fare,q= 4,labels = False)
df1
```

## Question 7

```python
df1 = df1.drop(["Cabin", "Name", "PassengerId", "Ticket", "Age", "Fare", "SibSp", "Parch" ],axis = 1)
df1
```

## Question 8

```python
df1 = pd.get_dummies(df1)
df1.head()
```

## Question 9

```python
X = df1.drop(["Survived"],axis = 1)
y = df1["Survived"]
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test = train_test_split(X,y,test_size = 0.3)
```

## Question 10

```python
from sklearn.linear_model import LogisticRegression
reg = LogisticRegression()
reg.fit(X_train,y_train)
```

```python
pred = reg.predict(X_test)
```

```python
result = pd.DataFrame({"Predicted":pred,"Actual":y_test})
result
```
